```python
from pypdf import PdfReader

def extract_text_and_pages(pdf_path):
    reader = PdfReader(pdf_path)
    pages = []
    full_texts = []
    for i, page in enumerate(reader.pages):
        txt = page.extract_text() or ""
        pages.append((i+1, txt))
        full_texts.append(txt)
    return "\n".join(full_texts), pages

full_text, pages = extract_text_and_pages(pdf_path)

len(pages), pages[0][1][:500]
```

```python
from nltk.tokenize import sent_tokenize

sentences = []
page_nums = []

for page_num, page_text in pages:
    sents = sent_tokenize(page_text)
    for s in sents:
        s_clean = s.strip()
        if s_clean:
            sentences.append(s_clean)
            page_nums.append(page_num)

len(sentences)
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

URL_RE = re.compile(r"https?://\S+|www\.\S+")

def preprocess(text):
    text = text.lower()
    text = URL_RE.sub(" ", text)
    token_pattern = r"\w+|[^\w\s]"
    tokens = re.findall(token_pattern, text)

    stopset = set(stopwords.words("english"))
    lem = WordNetLemmatizer()

    out = []
    for t in tokens:
        if re.fullmatch(r"[^\w\s]", t) or re.fullmatch(r"\d[\d\w]*", t):
            out.append(t)
            continue

        if t in stopset:
            continue

        out.append(lem.lemmatize(t))
    return " ".join(out)

preprocessed = [preprocess(s) for s in sentences]
preprocessed[:5]
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.preprocessing import normalize
import numpy as np

model = SentenceTransformer("all-MiniLM-L6-v2")

orig_emb = model.encode(preprocessed, convert_to_numpy=True, show_progress_bar=True)
orig_emb = normalize(orig_emb, axis=1)

orig_emb.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=0.95, random_state=42)
pca_emb = pca.fit_transform(orig_emb)

pca_emb.shape
```

```python
from umap import UMAP

umap = UMAP(
    n_components=15,
    n_neighbors=15,
    min_dist=0.0,
    metric="cosine",
    random_state=42
)

umap_emb = umap.fit_transform(pca_emb)
umap_emb.shape
```

```python
from hdbscan import HDBSCAN

hdb = HDBSCAN(
    min_cluster_size=8,
    metric="euclidean",
    prediction_data=True
)

labels = hdb.fit_predict(umap_emb)

labels[:20], len(set(labels))
```

```python
from bertopic import BERTopic

topic_model = BERTopic(
    embedding_model=None,
    umap_model=None,
    hdbscan_model=hdb,
    vectorizer_model=None,
    representation_model=None,
    nr_topics=None,
    calculate_probabilities=True
)

topics, probs = topic_model.fit_transform(
    preprocessed,
    embeddings=umap_emb,
)

topics[:20]
```

```python
topic_keywords = {tid: topic_model.get_topic(tid)
                  for tid in set(topics) if tid != -1}
topic_keywords
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import silhouette_score
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora import Dictionary

# Intra-cluster cosine
intra = {}
for u in set(topics):
    if u == -1:
        continue
    idx = np.where(topics == u)[0]
    if len(idx) < 2:
        intra[u] = None
        continue
    sims = cosine_similarity(orig_emb[idx])
    intra[u] = float(np.mean(sims[np.triu_indices_from(sims, k=1)]))

# NN-overlap
nn = NearestNeighbors(n_neighbors=10).fit(orig_emb)
neighbors = nn.kneighbors(orig_emb, return_distance=False)
overlaps = []
for i, neigh in enumerate(neighbors):
    same = set(np.where(topics == topics[i])[0])
    overlaps.append(len(same.intersection(neigh)) / 10)
nn_overlap = float(np.mean(overlaps))

# Silhouette
sil = float(silhouette_score(umap_emb, topics, metric="euclidean"))

# Topic coherence (c_v)
tokenized = [s.split() for s in preprocessed]
dictionary = Dictionary(tokenized)
topic_words = [[w for w, _ in topic_keywords[tid]] for tid in topic_keywords]
cm = CoherenceModel(topics=topic_words, texts=tokenized, dictionary=dictionary, coherence="c_v")
coh = cm.get_coherence()

# Page consistency
page_consistency = {}
pages_arr = np.array(page_nums)
for u in set(topics):
    if u == -1:
        continue
    idx = np.where(topics == u)[0]
    p = pages_arr[idx]
    diffs = []
    for i in range(len(p)):
        for j in range(i+1, len(p)):
            diffs.append(abs(p[i] - p[j]))
    page_consistency[u] = np.mean(diffs) if diffs else 0.0

intra, nn_overlap, sil, coh, page_consistency
```

```python
# -------------------------------
# 1. Extract raw text using pdfplumber
# -------------------------------

pages = []
with pdfplumber.open(PDF_PATH) as pdf:

    # Define 0-index page ranges
    selected_page_indices = (
        list(range(2, 30)) +      # pages 3–30
        list(range(48, 50)) +     # pages 49–50
        list(range(65, 68)) +     # pages 66–68
        list(range(69, 71))       # pages 70–71
    )

    for idx in selected_page_indices:
        if idx < len(pdf.pages):
            text = pdf.pages[idx].extract_text() or ""
            pages.append(text)

raw = "\n".join(pages)
```

```python
# -------------------------------
# 2. Normalize Unicode + whitespace
# -------------------------------
def normalize(text):
    text = unicodedata.normalize("NFKC", text)
    text = re.sub(r"\r\n?", "\n", text)
    text = re.sub(r"[ \t]+", " ", text)
    return text.strip()

raw = normalize(raw)
```

```python
# -------------------------------
# 4. Fix line breaks (VERY important for PDFs)
#    If a line ends with no punctuation, join it with next line.
# -------------------------------
processed_lines = []
lines = raw.split("\n")

buffer = ""
for line in lines:
    stripped = line.strip()
    if not stripped:
        continue

    if buffer == "":
        buffer = stripped
        continue

    # If previous line doesn't end with punctuation and current starts lowercase/digit
    if (not re.search(r"[.!?]$", buffer)) and re.match(r"^[a-z0-9]", stripped):
        buffer += " " + stripped
    else:
        processed_lines.append(buffer)
        buffer = stripped

if buffer:
    processed_lines.append(buffer)

clean_text = "\n".join(processed_lines)
```

```python

# -------------------------------
# 5. Sentence tokenization (NLTK)
# -------------------------------
sentences = sent_tokenize(clean_text)
```

```python
def remove_bullets(s):

    # Remove bullets at the START of sentence
    s = re.sub(r"^\s*[\•\-\*]\s*", "", s)                          # • bullet
    s = re.sub(r"^\s*[a-zA-Z]\s*[\.\)]\s*", "", s)                # a. , a)
    s = re.sub(r"^\s*\(?[a-zA-Z]\)\s*", "", s)                    # (a)
    s = re.sub(r"^\s*\(?[ivxIVX]+\)\s*", "", s)                   # (i), (iv)
    s = re.sub(r"^\s*[ivxIVX]+\s*[\.\)]\s*", "", s)               # i. , iv)

    # Remove bullets appearing IN THE MIDDLE of sentences
    s = re.sub(r"\s+[a-zA-Z]\s*[\.\)]\s+", " ", s)                # a. b. c.
    s = re.sub(r"\s+\(?[a-zA-Z]\)\s+", " ", s)                    # (a) (b)
    s = re.sub(r"\s+[ivxIVX]+\s*[\.\)]\s+", " ", s)               # i. ii. iii.
    s = re.sub(r"\s+\(?[ivxIVX]+\)\s+", " ", s)                   # (i) (ii)

    return s.strip()
```

```python
# Add this helper cleanup function
def clean_text(s):
    # Remove URLs
    s = re.sub(r"http\S+|www\.\S+", " ", s)

    # Remove tokens with digits (alphanumeric like A12, loan3, 2020-21)
    s = re.sub(r"\b[A-Za-z]*\d+[A-Za-z\d]*\b", " ", s)

    # Remove standalone numbers
    s = re.sub(r"\b\d+\b", " ", s)

    # Remove single-character words (exclude allowed acronyms)
    s = re.sub(r"\b\w\b", " ", s)

    # Collapse spaces
    s = re.sub(r"\s+", " ", s)
    return s.strip()


# OPTIONAL: Protect domain acronyms so they do NOT get removed
ACRONYM_WHITELIST = {"SLBC","DCC","LDM","MSME","PMFBY","NPA","SHG","KCC","NRLM","NULM","PMEGP"}

def restore_whitelisted_acronyms(text):
    words = text.split()
    new_words = []
    for w in words:
        if w.upper() in ACRONYM_WHITELIST:
            new_words.append(w.upper())  # restore domain acronym
        else:
            new_words.append(w)
    return " ".join(new_words)


def clean_punctuation(text):
    # 1) Normalize repeated punctuation
    text = re.sub(r"\.{2,}", ".", text)   # multiple dots -> single dot
    text = re.sub(r"\?{2,}", "?", text)   # multiple question marks -> ?
    text = re.sub(r"\!{2,}", "!", text)   # multiple exclamation marks -> !

    # 2) Remove ALL other punctuation except . ? !
    # Keep only: letters, digits, space, ., ?, !
    # Remove everything else
    text = re.sub(r"[^A-Za-z0-9\s\.\?\!]", " ", text)

    # 3) Collapse spaces
    text = re.sub(r"\s+", " ", text)
    return text.strip()


# -------------------------------
# FINAL CLEANING + FILTERING LOOP (UPDATED)
# -------------------------------
processed_sents = []

for s in sentences:
    s = s.strip()

    # Remove section numbers like 2.2.4, 1.1.3, etc.
    s = re.sub(r"^\d+(\.\d+){0,3}\s*", "", s)

    # Remove bullets (your old code)
    s = remove_bullets(s)

    # Remove URLs, alphanumeric tokens, numbers, single-letter words
    s = clean_text(s)

    # Remove punctuation except . ? !
    s = clean_punctuation(s)

    # Optionally restore whitelisted acronyms (recommended for RBI docs)
    s = restore_whitelisted_acronyms(s)

    # Final collapse
    s = re.sub(r"\s+", " ", s).strip()

    s = s.lower()

    if len(s) >= 2:  # prevent empty leftovers
        processed_sents.append(s)
```

```python
# 5) Basic embedding
model = SentenceTransformer("all-MiniLM-L6-v2")
embeddings = model.encode(text_sen, convert_to_numpy=True)
print("Embeddings shape:", embeddings.shape)
```

```python
pca = PCA(n_components=0.95, whiten=True, random_state=42)
embeddings_pca = pca.fit_transform(embeddings)
print("Embeddings shape after PCA:", embeddings_pca.shape)
```

```python
from sklearn.preprocessing import normalize

# Normalize original and PCA (PCA can be lower-dim)
orig_norm = normalize(embeddings, norm='l2', axis=1)         # (N, 384)
pca_norm  = normalize(embeddings_pca, norm='l2', axis=1)     # (N, D)

# Pairwise cosine matrices (both NxN)
orig_cos = orig_norm @ orig_norm.T
pca_cos  = pca_norm @ pca_norm.T

diff = orig_cos - pca_cos
abs_diff = np.abs(diff)

print("Mean abs difference (pairwise):", abs_diff.mean())
print("Median abs difference:", np.median(abs_diff))
print("Max abs difference:", abs_diff.max())

# Per-sentence average distortion
row_distortion = abs_diff.mean(axis=1)
worst = np.argsort(-row_distortion)[:10]
for i in worst:
    print(f"[{i}] avg_pairwise_distortion={row_distortion[i]:.4f} -> {sentences[i][:200]}...")
```

```python
import numpy as np
from sklearn.preprocessing import normalize

# embeddings: original (N, 384)
# embeddings_pca: PCA-transformed (N, D)
# pca: the fitted PCA instance

# 1) reconstruct PCA back to original space
emb_recon = pca.inverse_transform(embeddings_pca)   # shape (N, 384)

# 2) per-sentence reconstruction cosine
orig_norm = normalize(embeddings, axis=1)
recon_norm = normalize(emb_recon, axis=1)
cos_self = np.sum(orig_norm * recon_norm, axis=1)

# 3) pairwise structure distortion (if not computed already)
pca_norm = normalize(embeddings_pca, axis=1)
orig_cos = orig_norm @ orig_norm.T
pca_cos  = pca_norm @ pca_norm.T
abs_diff = np.abs(orig_cos - pca_cos)
row_distortion = abs_diff.mean(axis=1)

print("Mean self-recon cosine:", cos_self.mean())
print("Mean abs pairwise diff:", abs_diff.mean())

# show worst sentences by pairwise distortion and by recon-cos
worst_by_row = np.argsort(-row_distortion)[:20]
worst_by_recon = np.argsort(cos_self)[:20]

print("\nTop pairwise-distorted indices:")
for i in worst_by_row[:10]:
    print(i, f"row_dist={row_distortion[i]:.4f} recon_cos={cos_self[i]:.4f}", sentences[i][:200].replace("\n"," "))

print("\nTop low-reconstruction indices:")
for i in worst_by_recon[:10]:
    print(i, f"recon_cos={cos_self[i]:.4f} row_dist={row_distortion[i]:.4f}", sentences[i][:200].replace("\n"," "))
```

```python
from bertopic import BERTopic
from bertopic.dimensionality import BaseDimensionalityReduction

empty_reduction_model = BaseDimensionalityReduction()
```

```python
# Build a custom HDBSCAN model for BERTopic
# -----------------------
hdbscan_model = HDBSCAN(min_cluster_size=5,    # tune this: 5-20 typical
              min_samples=1,        # lower => more clusters; raise for tighter clusters
              metric='euclidean',   # using Euclidean on normalized PCA approximates cosine
              cluster_selection_method='eom',
              prediction_data=True)
```

```python
vectorizer_model = CountVectorizer(
    stop_words="english",
    ngram_range=(1, 2),
    min_df=2,                 # ignore extremely rare tokens
    max_df=0.95,              # ignore tokens appearing in >95% docs
)
```

```python
ctfidf_model = ClassTfidfTransformer(
    reduce_frequent_words=True,   # prevents very common tokens dominating
)
```

```python
topic_model = BERTopic(
    vectorizer_model=vectorizer_model,
    ctfidf_model=ctfidf_model,
    hdbscan_model=hdbscan_model,
    embedding_model=None,
    umap_model=empty_reduction_model,
    calculate_probabilities=True,
    verbose=True
)
```

```python
topics, probs = topic_model.fit_transform(
    text_sen,
    embeddings=embeddings_pca_norm
)
```

```python
topic_model.get_topic_info()
```

```python
new_topics = topic_model.reduce_outliers(text_sen, topics, probabilities=probs, strategy="distributions")
```

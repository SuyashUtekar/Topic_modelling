```python
# Load English stopwords
stop_words = set(stopwords.words('english'))

# Add unnecessary words (can be expanded as needed)
extra_words = {
    'mr', 'people', 'would', 'year', 'said', 'say', 'also', 'wale', 'could',
    'one', 'two', 'three', 'many', 'much', 'may', 'might', 'like', 'even',
    'still', 'new', 'made', 'make', 'get', 'take', 'got', 'well'
}
stop_words = stop_words.union(extra_words)

lemmatizer = WordNetLemmatizer()

allowed_punct = {'.', '?', '!', ';', ':'}

def remove_words_with_digits(text):
    return ' '.join(word for word in text.split() if not any(char.isdigit() for char in word))

def clean_text(text):
    """
    Clean and normalize input text:
    - Remove URLs
    - Remove words containing digits
    - Lowercase
    - Remove stopwords & extra words
    - Lemmatize
    - Remove extra whitespaces
    """
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    text = remove_words_with_digits(text)
    text = re.sub(r'\w*\d\w*', '', text)
    text = text.lower()

    tokens = nltk.word_tokenize(text)
    cleaned_tokens = []
    for token in tokens:
        if token in allowed_punct:
            cleaned_tokens.append(token)
        elif token.isalpha() and token not in stop_words:
            cleaned_tokens.append(lemmatizer.lemmatize(token))

    cleaned_text = ' '.join(cleaned_tokens)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()
    return cleaned_text

mixed_text = clean_text(mixed_text)
```

## Doing UMAP

```python
# Initiate UMAP
#umap_model = UMAP(n_neighbors= 5, n_components= 20, min_dist=0.0, metric='cosine', random_state=100)
```

```python
def auto_umap_params(embeddings):
    n_samples, embed_dim = embeddings.shape

    if n_samples < 50:
        n_neighbors = max(2, n_samples // 3)
    elif n_samples < 500:
        n_neighbors = 15
    elif n_samples < 5000:
        n_neighbors = 30
    else:
        n_neighbors = 50

    n_components = int(min(max(embed_dim * 0.1, 5), 20))
    return n_neighbors, n_components

def create_umap_model(embeddings, random_state=42):
    n_neighbors, n_components = auto_umap_params(embeddings)
    print(f"🧭 Using UMAP params → n_neighbors={n_neighbors}, n_components={n_components}")

    return UMAP(
        n_neighbors=n_neighbors,
        n_components=n_components,
        min_dist=0.0,
        metric='cosine',
        random_state=random_state
    )

# Usage
umap_model = create_umap_model(embed_mixed)
```

```python
umap_embed_mixed = umap_model.fit_transform(embed_mixed)
```

# Step 3: Using PCA then BERTopic implementation (fine Tuned)

```python
import hdbscan
import math
import umap

def hdbscan_auto(docs, embeddings):
    n_docs = len(docs)
    dist_matrix = cosine_distances(embeddings)
    # median distance per point
    median_dist = np.median(dist_matrix, axis=1)
    # heuristic: min_cluster_size = number of points closer than median distance
    #min_cluster_size = max(2, int(np.percentile(median_dist, n_docs)))
    if n_docs < 50:
        min_cluster_size = max(2, n_docs // 3)
    elif n_docs < 500:
        min_cluster_size = 15
    elif n_docs < 5000:
        min_cluster_size = 30
    else:
        min_cluster_size = 50

    cluster_model = hdbscan.HDBSCAN(
        min_cluster_size=min_cluster_size,
        metric='euclidean',
        cluster_selection_method='eom',
        prediction_data=True
    )
    cluster_model.fit(embeddings)
    return cluster_model
```

```python
print(hdbscan_auto(mixed_sen, pca_embed_mixed))
```

```python
#Fine Tuning terms

# C-TF-IDF
ctfidf_model = ClassTfidfTransformer(
    reduce_frequent_words=True,
    bm25_weighting=True
)

#Key-BERT representation model
rep_model = KeyBERTInspired()
```

```python
# 6️⃣ Initialize BERTopic with custom clustering
topic_model = BERTopic(
    embedding_model=embedding_model,
    calculate_probabilities=True,
    verbose=True,
    hdbscan_model=hdbscan_auto(mixed_sen, pca_embed_mixed),
    umap_model=None,
    ctfidf_model=ctfidf_model,
    #representation_model=rep_model,
)
```

## Fine-Tuning with UMAP

```python
def auto_umap_params(embeddings):
    n_samples, embed_dim = embeddings.shape

    if n_samples < 50:
        n_neighbors = max(2, n_samples // 3)
    elif n_samples < 500:
        n_neighbors = 15
    elif n_samples < 5000:
        n_neighbors = 30
    else:
        n_neighbors = 50

    n_components = int(min(max(embed_dim * 0.2, 5), 50))
    return n_neighbors, n_components

def create_umap_model(embeddings, random_state=42):
    n_neighbors, n_components = auto_umap_params(embeddings)
    print(f"🧭 Using UMAP params → n_neighbors={n_neighbors}, n_components={n_components}")

    return UMAP(
        n_neighbors=n_neighbors,
        n_components=n_components,
        min_dist=0.0,
        metric='cosine',
        random_state=random_state
    )

# Usage
umap_model = create_umap_model(embed_mixed)

# 6️⃣ CountVectorizer with stopword removal + bigrams
#vectorizer_model = CountVectorizer(stop_words='english', max_df=0.8)

# C-TF-IDF
ctfidf_model = ClassTfidfTransformer(
    reduce_frequent_words=True,
    bm25_weighting=True
)

#Key-BERT representation model
rep_model = KeyBERTInspired()
```

```python
# 6️⃣ Initialize BERTopic with custom clustering
topic_model = BERTopic(
    embedding_model=embedding_model,
    calculate_probabilities=True,
    verbose=True,
    hdbscan_model=hdbscan_auto(mixed_sen, embed_mixed),
    umap_model=umap_model,
    ctfidf_model=ctfidf_model,
    representation_model=rep_model
)
```

```python
# 7️⃣ Fit BERTopic with precomputed embeddings
topics, probs = topic_model.fit_transform(mixed_sen, embeddings=embed_mixed)
```

```python
# 8️⃣ Inspect results
print("Document topics:", topics)
print("\nTopic info:")
topic_model.get_topic_info()
```
